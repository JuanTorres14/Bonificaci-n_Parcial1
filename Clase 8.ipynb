# Polos y Ceros

```python
import matplotlib.pyplot as plt
import matplotlib.patches as paches

from scipy.signal import tf2zpk

%matplotlib inline
```

$H(z) =  \frac{z^2-5z+6}{z^2-2.5z+1}$

```python
num = [1, -5, 6]
den = [1, -2.5, 1]
```

```python
z,p,k = tf2zpk(num,den)
unit_c = plt.Circle((0,0),1, fill=False, color='darkcyan', ls='--')

fig, ax = plt.subplots()
ax.add_patch(unit_c)
plt.plot(p.real,p.imag,'x')
plt.plot(z.real,z.imag,'o')
plt.show()
```

Del diagrama anterior se puede evidenciar que si se requiere que el sistema sea un sistema real (causal) entonces z debe ser mayor a ambos polos en este caso 2 y 0.5, adicionalmente si ambos son causales esto significa que el sistema sería inestable ya que ambos se interceptarían en z>2.

Si se requiere que sea un sistema anticausal entonces z debería ser menor a ambos y de igual manera sería un sistema inestable ya que se interceptarían en z<0.5. 

De la única manera en la cual el sistema puede ser estable es cuando el sistema se comporta como un sistema bilateral ya que 0.5<z<2. Lo cual contiene el círculo unitario.
