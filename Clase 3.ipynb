# Periodicidad

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import random

%matplotlib inline
```

```python
numero = 10 

x = np.random.randint(0, 9, numero)
print(f'Secuencia: {x}')
```

```python
division = np.arange(-int(numero/2.0), int(numero/2.0), 1)

plt.figure(figsize=(20,7))
plt.stem(division, x)
plt.title('Señal continua',fontsize=14)
plt.xlabel("Time",fontsize=14)
plt.ylabel("Amplitude",fontsize=14)
plt.grid()
plt.show()
```

Apartir de la anterior secuencia representada en la gráfica, se puede evidenciar que para este caso se tiene una señal que no es periódica y adicional no es par ni tampoco impar.
