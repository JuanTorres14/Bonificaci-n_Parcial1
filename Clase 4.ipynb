# Reflejo de secuencias

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import random

%matplotlib inline
```

```python
x = []
for i in range(0, 10):
    x.append(random.randint(0, 10))
t = np.arange(0, 10)
print(x)

plt.stem(t, x)
plt.xlabel('Muestras')
plt.ylabel('Amplitud')
plt.grid()
```

```python
x_inv = x[::-1]
print(x_inv)

t = np.arange(-10, 0)

plt.stem(t, x_inv)
plt.xlabel('Muestras')
plt.ylabel('Amplitud')
plt.grid()
```

El origen de las muestras se sitúa en 0, por ende, a partir del 0 se refleja.
