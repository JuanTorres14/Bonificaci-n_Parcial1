# Desplazamiento de secuencias 

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import random

%matplotlib inline
```

## 2h(n)

```python
hn = []
for i in range(0, 10):
    hn.append(random.randint(0, 10))
t = np.arange(0, 10)

plt.stem(t, hn)
plt.xlabel('Muestras')
plt.ylabel('Amplitud')
plt.grid()
```

## h(n-2)

```python
hn_desplazado = []
for k in range (len(hn) + 2): 
    if k < 2:
        hn_desplazado.append(0)
    else: 
        hn_desplazado.append(hn[k-2]/2)


t = np.arange(0, len(hn_desplazado))

plt.stem(t, hn_desplazado)
plt.xlabel('Muestras')
plt.ylabel('Amplitud')
plt.grid()
    
```

El anterior proceso muestra el desplazamiento de una señal y como varía su amplitud al multiplicarse por un escalar.
