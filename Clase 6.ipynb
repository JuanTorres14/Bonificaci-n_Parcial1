# Correlación

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import random

%matplotlib inline
```

```python
#Correlación
numero = 10

x = np.random.randint(0, 9, numero)
print(f'Secuencia1: {x}')

y = np.random.randint(0, 5, numero)
print(f'Secuencia2: {y}')

Rxy = np.correlate(x, y, mode='full')

t = np.arange(-int(len(Rxy)/2), int(len(Rxy)/2)+1, 1)


#Gráfica
plt.figure(figsize=(8,4))
plt.stem(t, Rxy)
plt.xlabel('Muestras')
plt.ylabel('Amplitud')
plt.title('Pulso rectangular')
plt.grid()
plt.show()
```

Anteriormente se realizó la correlación, los picos máximos definen donde la señales tienen similitud una señal con la otra.
