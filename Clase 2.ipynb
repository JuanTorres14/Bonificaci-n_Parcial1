# Canal de transmisión

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import random

%matplotlib inline
```

```python
f = 0.1
t = np.arange(-10, 10, 0.001)


señal = np.sin(2*np.pi*f*t)

plt.figure(figsize=(20,7))
plt.plot(t, señal)
plt.title('Señal continua',fontsize=14)
plt.xlabel("Time",fontsize=14)
plt.ylabel("Amplitude",fontsize=14)
plt.grid()
plt.show()
```

```python
señal_1 = señal/8

plt.figure(figsize=(20,7))
plt.plot(t, señal_1)
plt.title('Señal atenuada',fontsize=14)
plt.xlabel("Time",fontsize=14)
plt.ylabel("Amplitude",fontsize=14)
plt.grid()
plt.show()
```

```python
ruido = np.random.randn(len(señal_1))/20
señal_2 = señal_1 + ruido

plt.figure(figsize=(20,7))
plt.plot(t, señal_2)
plt.title('Señal con ruido',fontsize=14)
plt.xlabel("Time",fontsize=14)
plt.ylabel("Amplitude",fontsize=14)
plt.grid()
plt.show()
```

En el proceso anterior, se puede presenciar como se atenúa la señal y se le agrega ruido. Podemos ver como la amplitud de la señal se ve afectada y como finalmente termina distorcionándose por completo.
