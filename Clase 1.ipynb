# Frecuencia de Nyquist

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import random

%matplotlib inline
```

```python
f = 0.1
t = np.arange(-10, 10, 0.001)


señal = np.sin(2*np.pi*f*t)

plt.figure(figsize=(20,7))
plt.plot(t, señal)
plt.title('Señal continua',fontsize=14)
plt.xlabel("Time",fontsize=14)
plt.ylabel("Amplitude",fontsize=14)
plt.grid()
plt.show()
```

```python
f_nyquist = 2*f

t1 = np.arange(-10, 10, 1/f_nyquist)

señal_1 = np.sin(2*np.pi*f*t1)

plt.figure(figsize=(20,7))
plt.plot(t1, señal_1)
plt.title('Señal continua',fontsize=14)
plt.xlabel("Time",fontsize=14)
plt.ylabel("Amplitude",fontsize=14)
plt.grid()
plt.show()
```

```python
f_nyquist = 20*f

t1 = np.arange(-10, 10, 1/f_nyquist)

señal_1 = np.sin(2*np.pi*f*t1)

plt.figure(figsize=(20,7))
plt.plot(t1, señal_1)
plt.title('Señal continua',fontsize=14)
plt.xlabel("Time",fontsize=14)
plt.ylabel("Amplitude",fontsize=14)
plt.grid()
plt.show()
```

De lo anterior podemos evidenciar que a veces no es suficiente con solo tener fs = 2fmax, sino que tiene que ser mucho mayor para que la señal pueda verse reconstruida por completo. 

Además, al muestrear la señal seno cuando fs=2fmax la señal coincide con los puntos en 0, por ende la señal da 0.
